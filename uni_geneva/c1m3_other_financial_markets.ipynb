## Emerging markets: stocks
[Emerging markets](https://www.coursera.org/learn/understanding-financial-markets/lecture/Iaz60/emerging-markets-stocks) (EM) represent a quarter of the world capitalization and are characterized by high growth but small market capitalization compared to their GDP. The [market capitalization to GDP ratio](https://en.wikipedia.org/wiki/List_of_countries_by_stock_market_capitalization) is an indication of the developemnt level of the country's financial market and of the size of the firms. On the other hand for developed country the market capitalization to GDP ratio, aka [Buffett indicator](https://en.wikipedia.org/wiki/Buffett_indicator) can be used to estimate how expensive is the aggregated stock market. Not all the outstanding stocks are available are available in the market. A percentage of a company capital is owned by the management of the firm, by local government or other firms or financial institution. The percentage of a company shares that are available on the market is called free-float shares and this number is the capital used to create indices. Regulations and explicit and indirect barriers can limit investments from foreigners. Emerging market have often higher risk-return ratio compared to developed market but the tail of the ditribution is often not symmetric and the left tail can be fat. The main risk for emerging markets, or **country risk** is [political stability](https://data.worldbank.org/indicator/PV.PER.RNK). Other risks are currency risk, liquidity risk.

## Gold
[Gold](https://www.coursera.org/learn/understanding-financial-markets/lecture/pWHvn/gold-the-ultimate-currency-part-1) can serve as an hedge against

* A falling US dollar
* Rising inflation
* Political tension
* Stock market crashes
* The sudden loss of faith in a fiat currency

It has been used as a store of value, a mean of exchange by itself and as a the payable value of paper notes (gold standard). Before [Bretton Woods](https://en.wikipedia.org/wiki/Bretton_Woods_system) in 1971, the money supply of a country and the currency exchange was linked to the amount of gold reserve owned by that country. The price of gold went from 25 USD in 1971 to 1000 USD in 2016, that is in 45 years. We can compute the annualized return of gold using the formula

$$25(1 + r)^{45} = 1000$$

so that

$$r = (\frac{1000}{25})^{\frac{1}{45}} - 1$$

```python
gold_end = 1000
gold_start = 25
years = 45
year_return = (gold_end / gold_start) ** (1/45) - 1
print('Gold annualized return: {:.0f}%'.format(year_return * 100))
```

If we plot the time series of gold and the USD in the same period we can see that the movement of gold and the USD are inversely correlated. Investors often see gold as a safe investment during political tensions but it is even more a choice to hedge their investments against inflation. The price of gold is positevely correlated with inflation. This is because gold works as a stable alternative to a currency. The cost of owning gold is that it doesn't provide returns, except its increase in value, and it must be stored somewhere. Gold can also work as an hedge against losses in the equity market and against extreme unexpected events.

## Real estate

In terms of returns and risk, real estate is in between bonds and stocks. Real estate usually provides a hedge against inflation as when inflation is soaring real estate prices go up as well.

## Hedge fund: leverage and short selling
A [hedge fund](https://en.wikipedia.org/wiki/Hedge_fund) aims to beat the market, that is achieving a result better than an average portfolio, using **leaverage** and **short selling**, irrespective of market direction aka **absolute return**. The remunaration of an asset management firm is a fee on average of 2% of the capital and 20% of the returns. A fund manager invests its own capital on the same assets on which they invest their clients capital so that they bear the same risk. The term *hedge fund* and the core ideas were developed by [Alfred Winslow Jones](https://en.wikipedia.org/wiki/Alfred_Winslow_Jones) in 1949. The assumptions for his strategy were 

1. I can't predict what the market will do in the future
2. I can identify undervalued securities and overvalued securities

He thought that he could make a profit on these assumptions by
* Buying undervalued stocks
* Short-selling overvalued stocks

### Leverage
[Leverage](https://en.wikipedia.org/wiki/Leverage_(finance)) is a performance booster that works by borrowingg assets to invest. If an investor borrows 100% of his own funds and the investment has a return of 10% then the real return is doubled since the invested capital is twice the original one reduced by the cost of borrowing. The problem with this technique is that it works both ways: an investor can have higher profits that what his wealth would allow but also higher losses if the investment is not successful. 

### Short-selling
[Short-selling](https://en.wikipedia.org/wiki/Short_(finance)) means selling something that we do not own. It works by borrowing an asset from an investment bank and immediately selling it on the stock market. It works when the asset is overvalued and the hort-seller predicts that its value will decrease. The capital made by selling the asset is kept in a bank account at the investment bank. If the price of the asset does indeed decrease by a certain threshold the investor can close the operation by telling the investment bank to buy the asset to be returned at its lower current price and taking the difference as a profit. If the price of the asset increases instead of decreasing then the investment may lead to a loss. The investment bank will ask the investor to put the difference between the initial price and the current increased one. A short-selling position is riskier than a long-position because if the investor does not back his investment he will have to pay for the asset at its increased price. If the investor decides to back the investment he has to take more risk betting on the decrease of an asset whose price increased against his of her prediction.
