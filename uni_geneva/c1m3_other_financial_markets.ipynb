The topics of this module are emerging markets, commodities like gold, real estate, hedge funds.

## Emerging markets
Emerging markets represent a quarter of the world capitalization and are characterized by high growth but small market capitalization compared to their GDP. The [market capitalization to GDP ratio](https://en.wikipedia.org/wiki/List_of_countries_by_stock_market_capitalization) is an indication of the developemnt level of the country's financial market and of the size of the firms. On the other hand for developed country the market capitalization to GDP ratio, aka [Buffett indicator](https://en.wikipedia.org/wiki/Buffett_indicator) can be used to estimate how expensive is the aggregated stock market. Not all the outstanding stocks are available are available in the market. A percentage of a company capital is owned by the management of the firm, by local government or other firms or financial institution. The percentage of a company shares that are available on the market is called free-float shares and this number is the capital used to create indices. Regulations and explicit and indirect barriers can limit investments from foreigners. Emerging market have often higher risk-return ratio compared to developed market but the tail of the ditribution is often not symmetric and the left tail can be fat. The main risk for emerging markets, or **country risk** is [political stability](https://data.worldbank.org/indicator/PV.PER.RNK). Other risks are currency risk, liquidity risk.
