### How to calculate the performance of an investment over several periods
The first concept is that performance over a number of periods, e.g. years, cannot be simply added to get the total performance. If you lose 50% of your investment in year 1 and gain 50 % in year 2 you still have a loss. We can see this with an example: letâ€™s say this is the performance of your portfolio over the past 4 years:

* Year 1: -10%
* Year 2: -45%
* Year 3: +20%
* Year 4: +35%

By how much do you need to perform in Year 5 in order to retrieve your initial capital?
We start with capital $c_0$. After the first year, with performance $p_1 = - 10\%$, our capital is $c_1 = c_0(1 + p_1)$. After the 2nd year, with performance $p_2 = -45\%$, our capital is $c_2 = c_1(1 + p_2)$ and so on

* Year 1: $c_1 = c_0(1 + p_1)$
* Year 2: $c_2 = c_1(1 + p_2)$
* Year 3: $c_3 = c_2(1 + p_3)$
* Year 4: $c_4 = c_3(1 + p_4)$
* Year 5: $c_5 = c_4(1 + p_5)$

The constraint to answer the question is $c_5 = c_0$ or

$$c_0(1 + p_1)(1 + p_2)(1 + p_3)(1 + p_4)(1 + p_5) = c_0$$

and 

$$p_5 = \frac{1}{\prod_{k=1}^4(1 + p_k)} - 1$$

```python
import numpy as np
p = np.array([-0.10, -0.45, 0.20, 0.35])
perf_y = p + 1
perf_y
```

```python
product = 1
for i in range(0, 4):
    product *= perf_y[i]
```

```python
product
```

```python
p5 = 1 / product - 1
print('Perfomance required in year 5 to have the same value as at the beginning of year 1: {:.2f}'.format(p5))
```

## Basic concepts in finance
A return is the ratio between the difference of the price of an equity at time t2 and t1 and the price at t1

$$r = \frac{p(t_2) - p(t_1)}{p(t_1)}$$

Before an investment we have to look at the past of a company not just the last return, and also at its volatility, that is the variation of the price in time. There is a trade-off between return and risk: the higher the return the higher is the risk. One way to consider both is to look at the return-risk ratio, an indicator similar to the Sharpe ratio. The higher is the return-risk ratio the better. We may have the same ratio with different return and risk values, in that case the choise depends on the utility function of the investor, whether he or she is more or less risk averse. One other mathematical tool beyond mean return and volatility is the distribution of the past returns. We take the returns over a period, let's say a couple of years and we build the histogram of the returns. From the histogram we can see what is the probability, or frequency, of negative returns and the asimmetry, or skewness, of the distribution.
