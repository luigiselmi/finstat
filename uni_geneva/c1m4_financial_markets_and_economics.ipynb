Some believe that financial markets participants are fully rational and an asset price contains all the information about the future cash flow of a firm needed to make an investment decision but this belief has been proven false: humans are not alway rational and the information may not be shared to every market participant in full.   

## Central banks
A central bank palys a key role in a country's economy: it can stimulate an economy by easing the provision of credit as liquidity to a depressed economy through commercial banks that borrow from the central bank or by restricting credit to an overheated economy. It is well known that it's easier to restrict credit for a central bank than to stimulate a depressed economy. The goal of all central bank is to keep inflation low, e.g. 2%. The US central bank, [the FED](https://www.federalreserve.gov/) has the additional goal of pursuing maximum employment. The economy is cyclical. It starts from a monetary stimulation from the central bank, perhaps also from the government by fiscal incentives, and by technical innovation. The economic growth increases labour demand but also wages and prices of raw materials. After some times wages and prices may increase more than the productivity and inflation starts. The central bank intervenes to reduce inflation by increasing the interest rate that results in lower employment since salaries are not flexible. Investors that are able to forecast a slowdown of the economy can increase the weight of bonds in their portfolio since central banks will likely lower their interest rate to stimulate it. The Nobel prize winner for economiy, Jan Timbergen stated a principle according to which is not possible to reduce inflation and at the same time keep employment high by just reducing interest rate. In order to recover from the financial crisis of 2008-2009 the conventional tools of the central banks to stimulate the economy were not enough. After the interest rate has been set to zero it cannot be used anymore and other unconventional monetary policies must be implemented. The [unconventional policies](https://www.ecb.europa.eu/press/key/date/2009/html/sp090428.en.html) are set up to inject liquidity in the economy and are called **quantitative easing** (QE). It works by buying sovereign and corporate bonds from commercial banks to that they have liquidity to provide credit. The US FED bought 3.5 Trillions of assets. QE probably helped to avoid a crisis like in the 1929 and it has been good for most investors whose shares, bonds and property have increased their value. On the other hand it has increased income inequality since workers have not seen an increase in wages ([video](https://www.coursera.org/learn/understanding-financial-markets/lecture/P4edd/central-banks-unconventional-policies-ubs-guest-lecturer)).

## The link between the economy and the equity market
China is expected to grow at 6% and to represent 20% of the world GDP by 2050, against 12.9 of the US and 9.7 of Europe. A growing economy does not always mean an increase in assets returns, for example inflation must be taken into account. As a matter of fatc, China GDP grew 4 times more than the US but the [S&P500](https://finance.yahoo.com/quote/%5EGSPC/) index grew 10 times more than the [MSCI China index](https://finance.yahoo.com/quote/MCHI/). This happens when growth is due excessive credit and high debt. When there is growth we should invest in equity and when there is lack of growth is better to put more weight on bonds. Equities outperform bonds so an investor should monitor the market for a recovery and rebalance the portfolio from bonds to equities.
