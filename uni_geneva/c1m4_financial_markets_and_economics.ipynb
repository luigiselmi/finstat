## Central banks
A central bank palys a key role in a country's economy: it can stimulate an economy by easing the provision of credit as liquidity to a depressed economy through commercial banks that borrow from the central bank or by restricting credit to an overheated economy. It is well known that it's easier to restrict credit for a central bank than to stimulate a depressed economy. The goal of all central bank is to keep inflation low, e.g. 2%. The US central bank, [the FED](https://www.federalreserve.gov/) has the additional goal of pursuing maximum employment. The economy is cyclical. It starts from a monetary stimulation from the central bank, perhaps also from the government by fiscal incentives, and by technical innovation. The economic growth increases labour demand but also wages and prices of raw materials. After some times wages and prices may increase more than the productivity and inflation starts. The central bank intervenes to reduce inflation by increasing the interest rate that results in lower employment since salaries are not flexible. Investors that are able to forecast a slowdown of the economy can increase the weight of bonds in their portfolio since central banks will likely lower their interest rate to stimulate it. The Nobel prize winner for economiy, Jan Timbergen stated a principle according to which is not possible to reduce inflation and at the same time keep employment high by just reducing interest rate. In order to recover from the financial crisis of 2008-2009 the conventional tools of the central banks to stimulate the economy were not enough. After the interest rate has been set to zero it cannot be used anymore and other unconventional monetary policies must be implemented. The [unconventional policies](https://www.ecb.europa.eu/press/key/date/2009/html/sp090428.en.html) are set up to inject liquidity in the economy and are called **quantitative easing** (QE). It works by buying sovereign and corporate bonds from commercial banks to that they have liquidity to provide credit. The US FED bought 3.5 Trillions of assets. QE probably helped to avoid a crisis like in the 1929 and it has been good for most investors whose shares, bonds and property have increased their value. On the other hand it has increased income inequality since workers have not seen an increase in wages ([video](https://www.coursera.org/learn/understanding-financial-markets/lecture/P4edd/central-banks-unconventional-policies-ubs-guest-lecturer)).

## The link between the economy and the equity market
Equity markets perform best when an economy is growing as in many emerging markets. As an example, China is expected to grow at 6% per year and to represent 20% of the world GDP by 2050, against 12.9 of the US and 9.7 of Europe. A growing economy does not always mean an increase in assets returns, for example inflation must be taken into account. As a matter of fact, China GDP grew 4 times more than the US but the [S&P500](https://finance.yahoo.com/quote/%5EGSPC/) index grew 10 times more than the [MSCI China index](https://finance.yahoo.com/quote/MCHI/). This happens when growth is based on excess credit and high debt. When there is growth we should invest in equity and when there is lack of growth is better to put more weight on bonds. Equities outperform bonds so an investor should monitor the market for a recovery and rebalance the portfolio from bonds to equities. There are four scenarios that depend on the economy growth and inflation that can guide equity investors

1. Strong GDP growth and high inflation (overheating)
2. Low GDP growth and high inflation (stagflation)
3. Low GDP growth and low inflation (deflation)
4. Strong GDP growth and low inflation (goldilocks)

Quite obviously for equity market the goldilocks scenario with strong GDP growth and low inflation is the best one.

## The link between the economy and the bond market
While the main drivers of the equity market are GDP growth and inflation, the bond market is driven by the interest rate and, more specifically, short-term interest rate and long-term interest rate. Normally interest rates, or the **yield curve**, increases the longer the maurity. This is because the longer the maturity the higher is the credi risk of the asset. One more reason is that long-term bonds incorporate the expectation of future inflation. In less normal time the yield curve may decrease in the long-term, that is short-term interest rate may be higher than long-term interest rate. Short-term interest rates are driven by central banks. When inflation is high central banks increase the interest rate. When the economy is depressed they decrease the interest rate to stimulate the economy. Long-term interest rates normally depend on the bond market and follow the short-term interest rate. Central bank can influence long-term interest rate when they conduct unconventional monetary policies, such as the so-called **quantitative easing**, by buying sovereign and corporate bonds to inject liquidity in the financial market. The yield curve, or its slope, is an indicator of the economy. When the slope is positive short-term interest rate is lower than the long-term and this means the economy is performing well and there is no or low risk of rising inflation. On the other hand, when the slope of the yield curve is negative the short-term interest rate is higher than the long -term interest rate and it means the inflation is rising and the central bank is removing liquidity from the market to cool-off the economy. From an investor point of view, bonds outperform equities when there is a slowdown of the economy, that is an increase of the interest rate and a decreasing GDP. According to what we have discussed so far, we look again at the main scenarios for the economy to understand which is better for equities and which for bonds

1. Strong GDP growth and high inflation (overheating -> sell bonds)
2. Low GDP growth and high inflation (stagflation -> sell equities)
3. Low GDP growth and low inflation (deflation -> buy bonds)
4. Strong GDP growth and low inflation (goldilocks -> buy equities)

For example in the 3rd scenario, "deflation" it is likely to predict that central bank will reduce the interest rate in the future and an investor may have a good return by investing in bonds before the central bank take action as predicted. 

## How interest rates affect equity and bond portfolios
The value of equities depends on the future expected cash flow and interest rate as can be seen from the formula

$$NPV = \sum_{t=1}^T \frac{C_t}{(1 + r)^t} - C_0$$

where $C_t$ represents future expected dividends and $r$ is the interest rate. So if the future cash flow decreases, often because of a lower economic growth, the value of equities decreases. Another reason for low equity return is an increase of the interest rate. A similar formula can be used for bond prices BP

$$BP = \sum_{t=1}^T \frac{Coupon_t}{(1 + YTM)^t} + \frac{Principal}{(1 + YTM)^N}$$

In this formula we can see that when the yield to maturity (YTM) increases the value of the bonds in our portfolio falls. The future performance of a bond can be predicted from its starting yield. This is because if the yield of a bond bought today is high, the probability that the future interest rate, and the yield, will be lower is high, thereby the bond price in the portfolio will increase. On the other hand if we buy a bond with a low yield the probability that the future interest rate will increase is high thereby the bond price will decrease. The slope of the yield curve depends on the interest rate. If the central bank raises the interest rate the yield curve flattens, that is its slope decreases. For bonds it is important to remind the relation between its price and its **[duration](https://en.wikipedia.org/wiki/Duration_(finance))** D, that measures its sensitivity to a change in interest rate $\Delta r$

$$\Delta BP = -D \frac{\Delta r}{1 + r} \times 100$$

The duration D is calculated as the present-value weighted average of the times of the cash flows. The formula above is its first derivative called *modified duration* that can be used as a first approximation. The duration should be reduced when a increase in interest rate is expected while it should be increased when is expected a decrease in interest rate ([video](https://www.coursera.org/learn/understanding-financial-markets/lecture/FV592/how-interest-rates-affect-equity-and-bond-portfolios-part-2))
