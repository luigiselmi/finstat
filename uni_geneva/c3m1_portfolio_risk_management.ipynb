# Course 3 - Module 1: Portfolio and Risk Management
We start this module with two key concepts:
1. The risk of a portfolio of two or more uncorrelated assets is lower than the mean of the risk of each asset. This is the outcome of diversification.
2. There is no free lunch: in order to get higher returns the investor must bear higher risk
3. Asset allocation is divided in two: strategic and tactic. Strategic asset allocation is about our profile or our client profile. Tactical asset allocation is about our view of the financial markets. When assets change the expected return we may rebalance our portfolio towards the more profitable assets. We will not change our strategic allocation because of a change in the asset returns.
