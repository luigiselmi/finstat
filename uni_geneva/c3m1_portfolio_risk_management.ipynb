# Course 3 - Module 1: Portfolio and Risk Management
We start this module with two key concepts:
1. The risk of a portfolio of two or more uncorrelated assets is lower than the mean of the risk of each asset. This is the outcome of diversification.
2. There is no free lunch: in order to get higher returns the investor must bear higher risk

Asset allocation is divided in two: strategic and tactic. Strategic asset allocation is about our profile or our client profile. Tactical asset allocation is about our view of the financial markets. When assets change the expected return we may rebalance our portfolio towards the more profitable assets. We will not change our strategic allocation because of a change in the asset returns.

## Distribution of returns
([Video](https://www.coursera.org/learn/portfolio-risk-management/lecture/V7G0S/distribution-of-returns-graphical-representation)) An investor that wants to choose how to allocate wealth to some financial assets can compare for their expected return and their risk. We can calculate the value of these two variables from the time series of the assets prices. Let's say we have the monthly prices of Microsoft (MSFT) and IBM for the last 10 years. We calculate the return of each asset as the change of asset price in two consecutive periods, months in our example. The return will be positive if the last price is higher than the previous one and negative if the last price is lower than the previous one. 

### Return distributions of the two assets
We assume that the time series of the returns are stationary, that is there is no trend like for the prices. We also assume, thanks to the stationarity, that we can calculate the average and the standard deviation of the time series of the returns indipendently of the time that is in the sample space. The sample space of the return is unidimensional and we divide it in bins to build the histogram of the returns of the two assets. The histograms represent the frequency of a certain level of return. We should observe that the most frequent return is close to zero while high returns, both gains and losses, are less frequent. Two assets have usually different return distributions. From each distribution we are interested in its average value, or mean return, and how it is dispersed. The measure of the average value is called **expected return** and the measure of dispersion is called **standard deviation** and it is calculated as the sum of the distances of the returns from the mean value. An investor must decide his or her **risk-return trade off**.

### Dependence between two assets' retuns
Beyond the expected return and the standard deviation of the two assets we are also interested in their relationship. We can use a scatterplot of the two monthly returns. The data points are build as pairs of returns for each time step in the two time series of the returns. If there exists any relationship we should see the data points spread more of less close to a line. The slope of this imaginary line is called **correlation** and it can take a value in the interval of real numbers [-1, 1]. The three information, expected return, standard deviation and correlation are all calculated from past data. In order to use this information to forecast future value we may assume that the future price will have the same statistical characteristics or we might use additional information.
